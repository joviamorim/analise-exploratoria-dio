```python
# criando o DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head(5)
```

```python
# verificando os tipos de dados
df.dtypes
```

```python
# criando uma coluna para custo
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
df.head(5)
```

```python
# qual o custo total?
round(df["Custo"].sum(), 1)
```

```python
# criando a coluna lucro
df["Lucro"] = df["Valor Venda"] - df["Custo"]
df.sample(5)
```

```python
# qual o lucro total?
df["Lucro"].sum()
```

```python
df["Tempo Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
df.sample(3)
```

```python
round(df.groupby("Marca")["Tempo Envio"].mean(), 1)
```

```python
# verificando valores ausentes
df.isnull().sum()
```

### Como ver lucro por ano e por marca?

```python
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
# gráfico de total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().plot.barh(title="Total produtos vendidos", color = "red")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df_2009 = df[df["Data Venda"].dt.year == 2009]
df_2009.sample(4)
```

```python
# gráfico de lucro por mês em 2009
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro por mês")
plt.xlabel("Mês da venda")
plt.ylabel("Lucro");
```

```python
df.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro por marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal");
```
